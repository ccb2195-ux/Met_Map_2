```python
import re

# ─── Lookup tables ──────────────────────────────────────────────────────────────

TYPO_MAP = {
    "Unitted States": "United States",
    "Austia":         "Austria",
    "Tunesia":        "Tunisia",
    "Gemany":         "Germany",
    "Indiaal":        "India",
}

NAME_MAP = {
    # ── United States variants ──
    "America":                              "United States",
    "USA":                                  "United States",
    "U.S.A.":                               "United States",
    "US":                                   "United States",
    "United States of America":             "United States",
    # ── Iran ──
    "Persia":                               "Iran",
    # ── Myanmar ──
    "Burma":                                "Myanmar",
    # ── Netherlands ──
    "Holland":                              "Netherlands",
    "The Netherlands":                      "Netherlands",
    "the Netherlands":                      "Netherlands",
    # ── Democratic Republic of the Congo ──
    "DRC":                                  "Democratic Republic of the Congo",
    "Dem. Rep. Congo":                      "Democratic Republic of the Congo",
    "Dem. Rep. of Congo":                   "Democratic Republic of the Congo",
    "Democratic Republic of Congo":         "Democratic Republic of the Congo",
    # ── Côte d'Ivoire ──
    "Ivory Coast":                          "Côte d'Ivoire",
    "Cote d'Ivoire":                        "Côte d'Ivoire",
    # ── "Republic of X" → common name ──
    "Republic of Benin":                    "Benin",
    "Republic of Cameroon":                 "Cameroon",
    "Republic of  Cameroon":                "Cameroon",
    "Republic of the Philippines":          "Philippines",
    "Republic of Costa Rica":               "Costa Rica",
    "Republic of Guinea":                   "Guinea",
    "Republic of Guinea-Bissau":            "Guinea-Bissau",
    "Republic of Palau":                    "Palau",
    "Republic of the Marshall Islands":     "Marshall Islands",
    "Republic of Kiribati":                 "Kiribati",
    "Republic of Timor-Leste":              "Timor-Leste",
    "Democratic Republic of Timor-Leste":   "Timor-Leste",
    "Republic of Madagascar":               "Madagascar",
    "the Republic of Congo":                "Republic of the Congo",
    # ── Other ──
    "Surinam":                              "Suriname",
    "Guinea Bissau":                        "Guinea-Bissau",
    "Russian Federation":                   "Russia",
    "Russia Federation":                    "Russia",
    "Aotearoa":                             "New Zealand",

    # ── Historical → modern: uncomment to enable, leave commented to preserve ──
    # "Gaul":           "France",
    # "New Spain":      "Mexico",
    # "Bohemia":        "Czech Republic",
    # "Byzantium":      "Turkey",
    # "Constantinople": "Turkey",
    # "Byzantine Egypt":"Egypt",
    # "Flanders":       "Belgium",
}

# Compound names containing "and" that must NOT be split
COMPOUND_COUNTRIES = {
    "Bosnia and Herzegovina",
    "Trinidad and Tobago",
    "Antigua and Barbuda",
    "Saint Kitts and Nevis",
    "Sao Tome and Principe",
    "São Tomé and Príncipe",
    "Wallis and Futuna Territory",
}

# ─── Cleaning function ──────────────────────────────────────────────────────────

def clean_country(raw):
    """
    Returns (country_clean, country_confidence).
    confidence: 'certain' | 'probable' | 'possible' | 'ambiguous'
    """
    if pd.isna(raw) or str(raw).strip() == '':
        return None, None

    text = str(raw).strip()
    confidence = 'certain'

    # 1. Fix known typos first
    if text in TYPO_MAP:
        text = TYPO_MAP[text]

    # 2. Normalise all-lowercase entries ("iran" -> "Iran", "india" -> "India")
    if text == text.lower() and len(text) > 1:
        text = text.title()

    # 3. Strip square-bracket annotations e.g. "present-day Turkey [Cover]" -> "present-day Turkey"
    text = re.sub(r'\s*\[[^\]]*\]', '', text).strip()

    # 4. Strip ":N.A." / ":N." annotations e.g. "Canada: N.A.", "France:N."
    text = re.sub(r'\s*:?\s*N\.?\s*(?:A\.?)?\s*$', '', text, flags=re.IGNORECASE).strip()

    # 5. Strip leading "from (the) " e.g. "from Albania", "from the Balkans"
    text = re.sub(r'^from\s+(?:the\s+)?', '', text, flags=re.IGNORECASE).strip()

    # 6. Strip "present-day" / "modern-day" prefix
    #    -? makes the hyphen optional, so both "present-day" and "present day" match
    text = re.sub(r'^(?:present-?day|modern-?day)\s+', '', text, flags=re.IGNORECASE).strip()

    # 7. Handle parentheticals — order matters here

    # (?) or (??) → mark possible and remove
    if re.search(r'\(\s*\?\??\s*\)', text):
        if confidence == 'certain':
            confidence = 'possible'
        text = re.sub(r'\s*\(\s*\?\??\s*\)', '', text).strip()

    # (probably) / (possibly) inside parens → mark confidence and remove
    m = re.search(r'\(\s*(probably|possibly|likely|perhaps)\s*\)', text, re.IGNORECASE)
    if m:
        q = m.group(1).lower()
        if confidence == 'certain':
            confidence = 'probable' if q == 'probably' else 'possible'
        text = re.sub(r'\s*\(\s*(?:probably|possibly|likely|perhaps)\s*\)', '', text, flags=re.IGNORECASE).strip()

    # (formerly X) / (former X) / (now X) / (modern X) → informational, remove
    text = re.sub(
        r'\s*\((?:formerly|former|now|modern|also|called|formerly known as)\s+[^)]+\)',
        '', text, flags=re.IGNORECASE
    ).strip()

    # Any remaining parenthetical → remove
    # This catches things like "(woven)", "(mounts)", "(cartoon)", "(British repro)"
    text = re.sub(r'\s*\([^)]*\)', '', text).strip()

    # 8. Multi-country separators → take first entry, mark ambiguous

    # Pipe: "Egypt|Egypt" or "France|Germany"
    if '|' in text:
        parts = list(dict.fromkeys([p.strip() for p in text.split('|') if p.strip()]))
        if len(parts) > 1:
            confidence = 'ambiguous'
        text = parts[0]

    # Slash: "Uganda/Tanzania", "Austria / Germany"
    if '/' in text:
        parts = [p.strip() for p in text.split('/') if p.strip()]
        if len(parts) > 1:
            confidence = 'ambiguous'
        text = parts[0]

    # Ampersand: "Republic of Congo & Angola"
    if '&' in text:
        parts = [p.strip() for p in text.split('&') if p.strip()]
        if len(parts) > 1:
            confidence = 'ambiguous'
        text = parts[0]

    # Strip trailing punctuation (commas, periods, colons) before checking comma lists
    text = re.sub(r'[,.\s:]+$', '', text).strip()

    # Comma list: "Bolivia, Peru, Chile" → "Bolivia"
    if ',' in text:
        parts = [p.strip() for p in text.split(',') if p.strip()]
        if len(parts) > 1:
            confidence = 'ambiguous'
        text = parts[0]

    # "and" separator — but skip known compound country names like "Bosnia and Herzegovina"
    if not any(text.lower() == c.lower() for c in COMPOUND_COUNTRIES):
        m = re.match(r'^(.+?)\s+and\s+.+$', text, re.IGNORECASE)
        if m:
            text = m.group(1).strip()
            confidence = 'ambiguous'

    # Strip dangling "or" / "and" left at end of string (e.g. "Spain or")
    text = re.sub(r'\s+(?:or|and)\s*$', '', text, flags=re.IGNORECASE).strip()

    # "or" separator: "Costa Rica or Panama"
    m = re.match(r'^(.+?)\s+or\s+.+$', text, re.IGNORECASE)
    if m:
        text = m.group(1).strip()
        confidence = 'ambiguous'

    # 9. Probability prefix: "probably Iran"
    m = re.match(r'^(probably|possibly|likely|perhaps)\s+(.+)$', text, re.IGNORECASE)
    if m:
        q = m.group(1).lower()
        text = m.group(2).strip()
        if confidence == 'certain':
            confidence = 'probable' if q == 'probably' else 'possible'

    # 10. Trailing uncertainty word: "Egypt possibly"
    m = re.match(r'^(.+?)\s+(possibly|probably|likely|perhaps)$', text, re.IGNORECASE)
    if m:
        text = m.group(1).strip()
        if confidence == 'certain':
            confidence = 'possible'

    # 11. Trailing question mark: "France?"
    if text.endswith('?'):
        text = text[:-1].strip()
        if confidence == 'certain':
            confidence = 'possible'

    # 12. Extended directional prefix: "Northwestern Iran" → "Iran"
    #     Covers: North/South/East/West + ern/ern/eastern/western/etc.
    #     Also: West-Central, South Central, Lower, Upper, Coastal
    m = re.match(
        r'^(?:North(?:east(?:ern?)?|west(?:ern?)?|ern?)?'
        r'|South(?:east(?:ern?)?|west(?:ern?)?|ern?)?'
        r'|East(?:ern?)?|West(?:ern?)?'
        r'|(?:West|East|South|North)[- ]?Central'
        r'|Central|Lower|Upper|Coastal|Outer)\s+(.+)$',
        text, re.IGNORECASE
    )
    if m:
        text = m.group(1).strip()  # confidence stays 'certain' — still a real country

    # 13. Final name standardisation map
    text = NAME_MAP.get(text, text)

    return text, confidence


# ─── Apply ─────────────────────────────────────────────────────────────────────

df[['country_clean', 'country_confidence']] = df['Country'].apply(
    lambda x: pd.Series(clean_country(x))
)

print("Confidence breakdown:")
print(df['country_confidence'].value_counts(dropna=False))
print(f"\nUnique clean countries: {df['country_clean'].nunique()}")
print("\nSample of cleaned values:")
df[df['Country'].notna()][['Country', 'country_clean', 'country_confidence']].drop_duplicates('Country').head(30)
```

```python
df_with_country = df[df['country_clean'].notna()]

df_with_country.to_csv("MetObjects_with_country.csv", index=False)

print(f"Saved {len(df_with_country):,} rows to MetObjects_with_country.csv")
```
